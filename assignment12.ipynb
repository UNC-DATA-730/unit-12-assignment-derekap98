**Create and visualize your model below...**

````r
library(recipes)
````

````r
Auto$displacement2 <- Auto$displacement ^2
rec <- lm(mpg ~ displacement + displacement2 + cylinders, data=Auto)
mod_fit <- rec %>% step_poly(displacement) %>% step_mutate(cyl_cat = factor(cylinders)) %>%
    step_dummy(cyl_cat) %>% step_rm(cylinders)

t <- predict(mod_fit)
ggplot(mod_fit, aes(x = displacement, y = mpg)) +
    geom_point(size = 1, alpha = 0.35) +
    geom_line(aes(y = t, color = factor(cylinders)), linewidth = 1)
````
